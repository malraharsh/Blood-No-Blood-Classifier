```python
import argparse
import os
import random

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import zipfile

from sklearn.metrics import (accuracy_score, classification_report,confusion_matrix)
from sklearn.model_selection import train_test_split
# from sklearn.preprocessing import LabelBinarizer
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.layers import (AveragePooling2D, Dense, Dropout, Flatten, Input, MaxPooling2D)
from tensorflow.keras.models import Model, load_model, Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import (ImageDataGenerator,img_to_array, load_img)
from tensorflow.keras.utils import to_categorical
```

```python
def get_model():
    base = MobileNetV2(input_tensor=Input(shape=(128,128, 3)), weights="imagenet", include_top=False)
    model = Sequential()
    model.add(base)
    model.add(MaxPooling2D(pool_size=(3, 3)))
    model.add(Flatten())
    model.add(Dense(128, activation="relu"))
    model.add(Dropout(0.4))
    model.add(Dense(2, activation="softmax"))

    base.trainable = False
            
    optimizer = Adam(lr=1e-4)
    model.compile(loss="binary_crossentropy", optimizer=optimizer, metrics=["accuracy"])
    
    return model
```

```python
model = get_model()
model.load_weights('Mobile-epochs-4-weights.h5')
```

```python
from utils import *
```

```python
data_dir = 'Data/Blood_noblood'
test_dir = os.path.join(data_dir, 'test')
```

```python
df_test = pd.DataFrame({'files': getFiles(test_dir), 'labels':'-1'})
```

```python
df_test.drop(index=2, inplace=True)
```

```python
BATCH_SIZE = 4
IMAGE_SIZE = (128, 128)
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = validation_datagen.flow_from_dataframe(
    df_test, 
    x_col='files',
    y_col='labels',
    target_size=IMAGE_SIZE,
    class_mode='categorical',
    batch_size=BATCH_SIZE
)
```

```python
preds = model.predict(test_generator, verbose=1)
preds = np.argmax(preds, axis=1)
```

```python
def show(df, preds):
    m = {1:'Blood', 0:'No-Blood'}
    plt.figure(figsize=(12, 12))
    for i in range(0, 2):
        plt.subplot(1, 2, i+1)
        image = plt.imread(df.iloc[i][0])
        plt.imshow(image)
        plt.title(m[preds[i]], color=('red' if preds[i] else 'blue'))
    plt.tight_layout()
    plt.show()
    
show(df_test, preds)
```
